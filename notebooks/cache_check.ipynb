```python
import joblib
import numpy as np
import glob
import os

def check_3d_failures(cache_dir="../data/processed"):
    # 1. Find the 3D cache file
    cache_files = glob.glob(os.path.join(cache_dir, "3d_descriptors_cache_*.joblib"))
    
    if not cache_files:
        print("No 3D cache files found in the directory.")
        return
        
    for cache_path in cache_files:
        print(f"\nAnalyzing: {os.path.basename(cache_path)}")
        
        # 2. Load the cached NumPy array
        X_3d = joblib.load(cache_path)
        
        # 3. Count the zero vectors
        # np.all(..., axis=1) checks if every single value in a row is exactly 0.0
        zero_mask = np.all(X_3d == 0, axis=1)
        num_failed = np.sum(zero_mask)
        total_mols = X_3d.shape[0]
        
        failure_rate = (num_failed / total_mols) * 100
        
        print(f"Total PROTACs processed: {total_mols}")
        print(f"Failed embeddings (Zero-Vectors): {num_failed}")
        print(f"Failure Rate: {failure_rate:.2f}%")

if __name__ == "__main__":
    check_3d_failures()
```
